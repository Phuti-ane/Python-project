```python
# @title
from bs4 import BeautifulSoup
import requests
import pandas as pd

#getting the url to be used to extract information
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_South_Africa'

#assign the link
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#we only want the table first before extracting exact data
table = soup.find_all('table')[0]
#table_text = table.text.strip()


#sorting in Heading(extracting headings)
table_title = table.find_all('th')
table_title_text = [title.text.strip() for title in table_title]

#put headings in a column with panda through data frame
df = pd.DataFrame(columns = table_title_text)
#create a 2 dimension table with panda
table_colu = table.find_all('tr')

for row in table_colu[1:]:
    row_data = row.find_all('td')
    individual = [row_text.text.strip() for row_text in row_data]
    length = len(df)
    df.loc[length] = individual
df
#we call df and not print
#===============================================================================================

#sorting out data
#select the first 10 rows and sort in profit and industry

df[df['Rank'].astype(int) < 10].sort_values( by= ['Profits(US$ millions)','Industry'])

#===============================================================================================

#setting multy indexes and sorting

df.set_index(['Rank','Company'], inplace = True)
df






```
